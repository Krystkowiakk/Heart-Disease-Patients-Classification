# Heart Disease Prediction 
## Metis Project 4 - Classification on Heart Disease Indicators (MVP)
Krystian Krystkowiak 2022

**Goal:**

This analysis seeks to establish a meaningfully predictive classification model for identifying high-risk patients. A model with acceptable accuracy, recall, precision, and F1 may be utilised for more informed outreach.

Heart diseased patients are (~8%) minority of the target distribution, thus we will priority importance of model recall for this specific problem.

**Process:**

The dataset originally comes from the CDC and is a major part of the Behavioural Risk Factor Surveillance System (BRFSS). It is initially cleaned by Kamil Pytlak, who shared it at Kaggle:
www.kaggle.com/datasets/kamilpytlak/personal-key-indicators-of-heart-disease

Data contains over 319k observations with 19 features and is a balanced mix of quantitative and qualitative features. Categorical features were binarized (using `get_dummies`) and numerical features were standardised for modeling.

**Preliminary conclusions:**

Classification analysis using Logistic Regression and Random Forest models suggests that balanced Logistic Regression model may have more utility in this case, presenting initially better recall and F-1 scores compared to the Random Forest (see Figures 1 and 2 below).

**From here:**

This analysis will be expanded by employing boosted tree models, as well as KNN. Oversampling and under-sampling of the target class may also improve model scoring metrics.
